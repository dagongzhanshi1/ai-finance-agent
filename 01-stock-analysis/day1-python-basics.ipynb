```python
def calculate_return(buy_price, sell_price):                                                                                                                                                              
         return (sell_price - buy_price) / buy_price * 100                                                                                                                                                     
                                                                                                                                                                                                               
print(f"收益率: {calculate_return(100, 120):.2f}%")    
```
